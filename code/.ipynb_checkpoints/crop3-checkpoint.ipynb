```python
import numpy as np # linear algebra
import xml.etree.ElementTree as ET # for parsing XML
import matplotlib.pyplot as plt # to show images
from PIL import Image # to read images
import os
import glob
```

```python
root_images="../input/all-dogs/pics/"
root_annots="../annotation/Annotation/"
```

```python
all_images=os.listdir(root_images)
print(f"Total images : {len(all_images)}")

breeds = glob.glob('../annotation/Annotation/*')
annotation=[]
for b in breeds:
    annotation+=glob.glob(b+"/*")
print(f"Total annotation : {len(annotation)}")

breed_map={}
for annot in annotation:
    breed=annot.split("/")[-2]
    index=breed.split("-")[0]
    breed_map.setdefault(index,breed)
    
print(f"Total Breeds : {len(breed_map)}")
```

```python
def bounding_box(image):
    bpath=root_annots+str(breed_map[image.split("_")[0]])+"/"+str(image.split(".")[0])
    tree = ET.parse(bpath)
    root = tree.getroot()
    objects = root.findall('object')
    
    im=Image.open(os.path.join(root_images,image))
    imsize = im.size
#     print(imsize)
    
    for o in objects:
        bndbox = o.find('bndbox') # reading bound box
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

#         if imsize[0] < imsize[1]:
#             xmin = 0
#             xmax = imsize[0]
#             if ymax - ymin < imsize[1]:
#                 if ymax - imsize[0] >= 0:
#                     ymin = ymax - imsize[0]
#                 else:
#                     ymin = 0
#                     ymax = imsize[0]
#             else:
#                 ymin = ymax - imsize[0]
#         else:
#             ymin = 0
#             ymax = imsize[1]
#             if xmax - xmin < imsize[0]:
#                 if xmax - imsize[1] >= 0:
#                     xmin = xmax - imsize[1]
#                 else:
#                     xmin = 0
#                     xmax = imsize[1]
#             else:
#                 xmin = xmax - imsize[1]

#         if imsize[0] < imsize[1]:
#             xmin = 0
#             xmax = imsize[0]
#             if ymax - ymin < imsize[1]:
#                 if ymin + imsize[0] <= imsize[1]:
#                     ymax = ymin + imsize[0]
#                 else:
#                     ymax = imsize[1]
#                     ymin = imsize[1] - imsize[0]
#             else:
#                 ymax = ymin + imsize[0]
#         else:
#             ymin = 0
#             ymax = imsize[1]
#             if xmax - xmin < imsize[0]:
#                 if xmin + imsize[1] <= imsize[0]:
#                     xmax = xmin + imsize[1]
#                 else:
#                     xmax = imsize[0]
#                     xmin = imsize[0] - imsize[1]
#             else:
#                 xmax = xmin + imsize[1]

        if imsize[0] < imsize[1]:
            xmin = 0
            xmax = imsize[0]
            ycenter = (ymax + ymin)/2 + 0.1
            
            if ycenter + imsize[0]/2 <= imsize[1]:
                if ycenter - imsize[0]/2 >= 0:
                    ymax = round(ycenter + imsize[0]/2)
                    ymin = round(ycenter - imsize[0]/2)
                else:
                    ymin = 0
                    ymax = imsize[0]
            else:
                ymax = imsize[1]
                ymin = ymax - imsize[0]
        else:
            ymin = 0
            ymax = imsize[1]
            xcenter = (xmax + xmin)/2 + 0.1
            
            if xcenter + imsize[1]/2 <= imsize[0]:
                if xcenter - imsize[1]/2 >= 0:
                    xmax = round(xcenter + imsize[1]/2)
                    xmin = round(xcenter - imsize[1]/2)
                else:
                    xmin = 0
                    xmax = imsize[1]
            else:
                xmax = imsize[0]
                xmin = xmax - imsize[1]
                
#         print(xmax-xmin, ymax-ymin)
    return (xmin,ymin,xmax,ymax)
```

```python
fig, axes = plt.subplots(5, 10, figsize=(15, 8),
                         subplot_kw={'xticks': (), 'yticks': ()})

for image, ax in zip(all_images, axes.ravel()):
    bbox=bounding_box(image)
    im=Image.open(os.path.join(root_images,image))
    im=im.crop(bbox)
    im = im.resize((64,64), Image.LANCZOS)
#     print(im.size)
    plt.axis("off")
    ax.imshow(im)
```

```python
for image in all_images:
    bbox=bounding_box(image)
    im=Image.open(os.path.join(root_images,image))
    im=im.crop(bbox)
    im = im.resize((64,64), Image.LANCZOS)
    savepath = "../input/all-dogs-crop3/pics/" + image
    im.save(savepath,"PNG")
```
