```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
imagepath1 = '../output_images/cropped_200epoch/'
```

```python
picnames = []
randompics = []
picnum = 50

for i in range(picnum):
    randomint = int(random.random()*10000)
    picnames.append('{:0=5}'.format(randomint))
    picname = imagepath1 + 'image_' + str('{:0=5}'.format(randomint)) + '.png'
    img = mpimg.imread(picname)
    randompics.append(img)
```

```python
fig, axes = plt.subplots(5, 10, figsize=(15, 8),
                         subplot_kw={'xticks': (), 'yticks': ()})

for image, ax in zip(randompics, axes.ravel()):
    ax.imshow(image)
```
