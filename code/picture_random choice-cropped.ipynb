```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
imagepath0 = '../input/all-dogs/pics/'
imagepath1 = '../input/all-dogs-crop/pics/'
imagepath2 = '../input/all-dogs-crop2-64/pics/'
```

```python
picnames = []
randompics = []
picnum = 50
imagepath = imagepath1

for i in range(picnum):
    file  = random.choice(os.listdir(imagepath))
    picname = imagepath + file
    img = mpimg.imread(picname)
    randompics.append(img)
    
    
fig, axes = plt.subplots(5, 10, figsize=(15, 8),
                         subplot_kw={'xticks': (), 'yticks': ()})

for image, ax in zip(randompics, axes.ravel()):
    ax.imshow(image)
```
